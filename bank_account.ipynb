```python
class BankAccount:
    ''' Bank Account management '''
    def __init__(self, name, date_of_birth,  account_number,  account_type, balance = 0):
        self.name = name
        self.date_of_birth = date_of_birth
        self.account_number = account_number
        self.account_type = account_type
        self.balance = balance
        self.filename = str(self.account_number)+"_"+self.account_type+"_"+self.name + ".txt"

    def creating_account(self):
        ''' Creating am account for a user '''

        print("Welcome to RBC Bank now you are goint to create a bank account")

        # getting the new user informatons 
        self.name = input("Enter your full name: ")
        self.date_of_birth = input("Enter your date of birth (DD/MM/YYYY): ")
        self.account_number = self.name[:2].upper() + self.date_of_birth.replace("/","")
        
        # choising the account type 
        account_type_choice = {
            "a": "Checking",
            "b": "Saving",
            "c": "Business"
        }
        for key in account_type_choice:
            print(f"({key}) : {account_type_choice[key]}")
        x = input("Your choice : ").lower()
        if x == "a":
            self.account_type = "Checking"
        elif x == "b":
            self.account_type = "Saving"
        elif x == "c":
            self.account_type ="Business"
        else:
            print("Enter a valid choice")

        # Summary of the account
        self.filename = f"{self.account_number}_{self.account_type}_{self.name}.txt"
        print(f"A {self.account_type} account for {self.name} had been created successfully with acount number : {self.account_number}")
    
    def deposit_money(self):
        ''' Deposit money and record the transaction. '''
        
        amount = int(input("Enter the amount you want to deposit : $"))
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        
        # Update balance
        self.balance += amount
        print(f"${amount} deposited successfully. New balance : ${self.balance}")

        # Record transaction in the statement file
        with open(self.filename, "a", encoding = "utf-8") as file:
            file.write(f"Deposit : +${amount} | New balance : ${self.balance}\n")

    def withdraw_money(self):
        ''' Withdrawing money from the account '''
        
        amount = int(input("Enter the amount you want to withdraw: $"))

        if amount > self.balance:
            print("insuficient amount in the account right now")
            return
        
        # Update the balance 
        self.balance -= amount
        print(f"${amount} has been withdraw from your account. New balance : ${self.balance}")

        # record the transaction in the statement file
        with open(self.filename, "a", encoding= "utf-8") as file:
            file.write(f"Withdraw : ${amount} | new balance : ${self.balance}\n")
    
    def checking_balance (self):
        ''' Checking the balance '''
        print(f" Your actual balance is : ${self.balance}")
    
    def checking_account_type(self):
        ''' Checking the account type '''
        print(f"Your account type is : {self.account_type}")

    def checking_account_number(self):
        ''' Checking the account number '''
        print(f"Your account number is : {self.account_number}")
    
    def checking_holder_name(self):
        ''' checking holder name '''
        print(f"The account belong to : {self.name}")
    
    def transaction_history(self):
        ''' Read the file and display all the transactions from the statement file. '''
        try :
            with open(self.filename, "r" , encoding= "utf-8") as file:
                print(f"\n --- Transaction history for {self.name} ---")
                for line in file:
                    print(line.strip())
        except FileNotFoundError:
            print("No transaction found yet")

def main():
    account = BankAccount("", "", "", "")
    account.creating_account()
    account.deposit_money()
    account.withdraw_money()
    account.checking_account_number()
    account.checking_account_type()
    account.checking_balance()
    account.checking_holder_name()
    account.transaction_history()

main()
```
